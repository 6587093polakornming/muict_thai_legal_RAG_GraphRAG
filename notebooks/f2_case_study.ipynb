```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path.cwd().parent
EVAL_DIR =  BASE_DIR / "data" / "evaluation" / "results"

# Load Data
hybrid_rag_path = EVAL_DIR / "hybrid_rag_topk2_results.csv"
hybrid_ref_rag_path = EVAL_DIR /  "hybrid_ref_topn_parent_topk2_results.csv"
graph_rag_path = EVAL_DIR / "graph_topk2_results.csv"
 
# Top k = 2
hybrid_rag_df = pd.read_csv(hybrid_rag_path) 
hybrid_ref_rag_df = pd.read_csv(hybrid_ref_rag_path) 
graph_rag_df = pd.read_csv(graph_rag_path) 
```

```python
hybrid_rag_df.head(1)
```

```python
from functools import reduce

# Scnario 1
# All Correct

# Multi-HitRate = 1
hybrid_rag_df_filter = hybrid_rag_df[hybrid_rag_df["multi_hit_rate"] == 1]
hybrid_ref_rag_df_filter = hybrid_ref_rag_df[hybrid_ref_rag_df["multi_hit_rate"] == 1]
graph_rag_filter = graph_rag_df[graph_rag_df["multi_hit_rate"] == 1]

# Intersection with id
dfs = [hybrid_rag_df_filter, hybrid_ref_rag_df_filter, graph_rag_filter]

# Intersect based on a common column named 'ID'
df_intersection = reduce(lambda left, right: pd.merge(left, right, on='id', how='inner'), dfs)
```
