# Nitibench Data Preparation

````python
import pandas as pd
````

### Data source

````python
from pathlib import Path

current_directory = Path.cwd()
data_dir = current_directory.parent / 'data' / 'raw' 
nitibench_path = data_dir / 'nitibench' / 'data' / 'nitibench-ccl.parquet'

print(f"Current Directory: {current_directory}")
print(f"Dataset Directory:    {nitibench_path}")
````

### Output Dir

````python
output_dir = current_directory.parent / 'data' / "processed" 
````

### Lookup Dataset

````python
df = pd.read_parquet(nitibench_path)
df.head(2)
````

````python
df.info()
````

use ```relevent_laws``` and ```reference laws``` as knowledge base in RAG

## De-Duplication

````python
# crete empty df
output_df = pd.DataFrame(columns=["law_name", "section_content" ,"section_num"])
````

````python
all_data = [item for col in ["relevant_laws", "reference_laws"] 
            for row in df[col] 
            for item in row]

output_df = pd.DataFrame(all_data).drop_duplicates(subset=["law_name", "section_num"])

print(f"rows: {len(output_df)}")
print(output_df.info())
````

# Export to Parquet

````python
from datetime import date

# Get today's date
today = date.today()

output_filepath = output_dir / f"nitibench_cleaned_{today}.parquet"
print(output_filepath)
````

````python
# Export the DataFrame to a Parquet file
output_df.to_parquet(output_filepath,  index=False)
````
